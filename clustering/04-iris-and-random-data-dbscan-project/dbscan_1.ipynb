```python
from sklearn.cluster import DBSCAN
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris=datasets.load_iris()
x=iris.data
x[0:5]
```

```python
model=DBSCAN(eps=1,min_samples=4)
model.fit(x)
```

```python
model.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c = model.labels_);
```
