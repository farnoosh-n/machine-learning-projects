## hierarchical Clustering on iris dataset

```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt
from sklearn import cluster,datasets
from scipy.cluster import hierarchy 
```

```python
iris=datasets.load_iris()
x=iris.data[:,:2]
```

```python
model1=cluster.AgglomerativeClustering(n_clusters=2,linkage='average')
model1.fit(x)
model2=cluster.AgglomerativeClustering(n_clusters=3,linkage='average')
model2.fit(x)
model3=cluster.AgglomerativeClustering(n_clusters=4,linkage='average')
model3.fit(x)
```

```python
plt.figure(figsize=(9,3))

plt.subplot(131)
plt.scatter(x[:,0],x[:,1],c=model1.labels_ )
plt.title("k=2")

plt.subplot(132)
plt.scatter(x[:,0],x[:,1],c=model2.labels_ )
plt.title("k=3")

plt.subplot(133)
plt.scatter(x[:,0],x[:,1],c=model3.labels_ )
plt.title("k=4")
```
