## Simple-Linear-Regression

```python
x2=np.asarray(df[['Area']])
y2=np.asarray(df['Price(USD)'])
y2[0:5]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

x2_train, x2_test, y2_train, y2_test = train_test_split( x2, y2, test_size=0.2, random_state=4)
print ('Train set:', x2_train.shape,  y2_train.shape)
print ('Test set:', x2_test.shape,  y2_test.shape)
```

```python
model = LinearRegression()
model.fit (x2_train, y2_train)
```

```python
y2_pred=model.predict(x2_test)
y2_pred[0:5]
```

```python
from sklearn.metrics import r2_score

print("R2-score: %.2f" % r2_score(y2_test , y2_pred) )
```

## NoneLinearRegression
### Polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

x3=np.asarray(df[['Area']])
y3=np.asarray(df['Price(USD)'])

x3_train, x3_test, y3_train, y3_test = train_test_split( x3, y3, test_size=0.2, random_state=4)
print ('Train set:', x3_train.shape,  y3_train.shape)
print ('Test set:', x3_test.shape,  y3_test.shape)

model = PolynomialFeatures(degree=2)
model.fit (x3_train, y3_train)

train_x_poly = model.fit_transform(x3_train)
train_x_poly
```

```python
clf = linear_model.LinearRegression()
train_y_ = clf.fit(train_x_poly, y3_train)
# The coefficients
print ('Coefficients: ', clf.coef_)
print ('Intercept: ',clf.intercept_)
```

```python
from sklearn.metrics import r2_score

test_x_poly = model.fit_transform(x3_test)
test_y_ = clf.predict(test_x_poly)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - y3_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - y3_test) ** 2))
print("R2-score: %.2f" % r2_score(y3_test,test_y_ ) )
```
